#### Type conversions - Renaming columns - Cleaning

`````julia
sdf1[!,:Year] = parse.(Int64, sdf1[!,:Year])  # using ! rather than : --> no need for copies
sdf1[!,"Disaster Subgroup"] = convert.(String, sdf1[!, "Disaster Subgroup"])
eltype.(eachcol(sdf1))
`````

`````julia
rename!(sdf1, "Disaster Subgroup" => "Subgroup")
`````

`````julia
levels(sdf1.Subgroup)
`````

`````julia
#sdf1[sdf1.Subgroup .== "Extra-terrestrial", :]
for sgr ∈ levels(sdf1.Subgroup)
    noccs = nrow(sdf1[sdf1.Subgroup .== sgr, :])
    println("$sgr : $noccs  occurrences")
end
`````

`````julia
nrow(sdf1)
`````

`````julia
sdf1 = sdf1[sdf1.Subgroup .!= "Extra-terrestrial", :]
`````

`````julia
nrow(sdf1)
`````

We are only interested in the years from 2000 to 2020

`````julia
sdf1 = sdf1[sdf1.Year .≥ 2000, :]
`````

### Groupy && Combine

`````julia
using Chain
res = @chain sdf1 begin
    groupby([:Year, :Subgroup])
    combine(nrow => :Disasters)
end
`````

`````julia
sort!(res, [:Year, :Subgroup])
`````

### Plotting

`````julia
using StatsPlots, GR
gr()

years = levels(res.Year)
subgroups = levels(res.Subgroup)
sgs = reshape(subgroups, 1, length(subgroups))
data = reshape(res.Disasters, length(subgroups), length(years))
groupedbar(years, data', bar_position = :stack, bar_width = 0.4, alpha= 0.5,
    labels=sgs,
    ylims=(0,650), xlabel="Year", ylabel="Disaster Subgroup", 
    #xticks=years, xrotation=45,
    legend=:topright, size=(600,400))
`````

## Total Deaths: what is the trend ?

`````julia
sdf2 = select(df1,"Year", "Disaster Subgroup", "Total Deaths")
`````

### Rename Columns 

`````julia
sdf2 = rename!(sdf2, "Disaster Subgroup" => "Subgroup", "Total Deaths" => "Deaths")
`````

### Drop  `missing` & Type conversions

`````julia
# check for missing values
missingn = nrow(sdf2) - sum(completecases(sdf2))
println("$missingn missing values for Deaths column.")
`````

`````julia
nrow(sdf2)
`````

`````julia
df = dropmissing(sdf2)
nrow(df)
`````

Alternatively, to keep the same numbers of disasters, we could replace the `missing` values by 0 or NaN instead of dropping these lines off with `dropmissing`.
````
replace!(sdf2.Deaths, missing => 0)
sdf2
````

`````julia
eltype.(eachcol(df))
`````

`````julia
df[!,:Year] = parse.(Int64, df[!,:Year])  # using ! rather than : --> no need for copies
df[!,:Subgroup] = convert.(String, df[!, :Subgroup])
df[!,:Deaths] = convert.(Int64, skipmissing(df[!,:Deaths]))
eltype.(eachcol(df))
`````

### Filtering
We are only interested in years from 2000

`````julia
levels(df.Subgroup) # Ok extra-terrestrial has already been removed !!!
`````

`````julia
df = df[df.Year .≥ 2000, :]
nrow(df)
`````

### Group by - Combine - Sort

`````julia
using Chain
res = @chain df begin
    groupby([:Year, :Subgroup])
    combine(:Deaths => sum => :Deaths)
    sort(_, [:Year, :Subgroup])
end
`````

### Plotting

`````julia
years = levels(res.Year)
subgroups = levels(res.Subgroup)
sgs = reshape(subgroups, 1, length(subgroups))
data = reshape(res.Deaths, length(subgroups), length(years))
groupedbar(years, data', bar_position = :stack, bar_width = 0.4, alpha= 0.5,
    labels=sgs,
    ylims=(0,400000), xlabel="Year", ylabel="Reported Deaths", 
    #xticks=years, xrotation=45,
    legend=:topright, size=(600,400))
`````
