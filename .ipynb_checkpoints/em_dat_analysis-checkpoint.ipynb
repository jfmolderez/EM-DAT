```julia
df0 = DataFrame(XLSX.readtable(
        "data/emdat_public_2021_07_30_query_uid-IszNK0.xlsx", 
        "emdat data",
        "A:AU";
        first_row = 7,
        header = true
        )...)
```

```julia
sdf = select(df0, "Year", "Disaster Subgroup")
```

```julia
sdf[!,:Year] = parse.(Int64, df0[!,:Year])
```

```julia
sdf[!,"Disaster Subgroup"] = convert.(String, df0[!, "Disaster Subgroup"])
```

```julia
sdf
```

```julia
disasters = sdf[sdf.Year .≥ 2000, :]
```

```julia
rename!(disasters, "Disaster Subgroup" => "Subgroup")
```

```julia
disasters = combine(groupby(disasters, [:Year, :Subgroup]), nrow => :Disasters)
```

```julia
levels(disasters.Subgroup)
```

```julia
disasters[disasters.Subgroup .== "Extra-terrestrial", :]
```

```julia
disasters = disasters[disasters.Subgroup .!= "Extra-terrestrial", :]
```

```julia
sort!(disasters, [:Year, :Subgroup])
```

```julia
years = levels(disasters.Year)
subgroups = levels(disasters.Subgroup)
data = reshape(disasters.Disasters, length(years), length(subgroups))
p = groupedbar(years, data, bar_position=:stack)
plot(p, bar_width=0.7, fill=[:blue :red :green :yellow :purple], 
    labels=["Biological" "Climatological" "GeoPhysical" "Hydrological" "Meteorological"],
    #labels=reshape(subgroups,length(subgroups),1),
    ylims=(0,600), xlabel="Year", ylabel="Disaster Subgroup", 
    xticks=years, xrotation=45,
    legend=:topleft, size=(600,400))
```
